### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
import seaborn as sns

facturacion = {nombre: df["Precio"].sum() for nombre, df in tiendas.items()}
facturacion_df = pd.DataFrame(list(facturacion.items()), columns=["Tienda", "Facturación"])

sns.barplot(data=facturacion_df, x="Tienda", y="Facturación")
plt.title("Facturación total por tienda")
plt.ylabel("Facturación en pesos")
plt.xticks(rotation=45)
plt.show()
```

# 2. Ventas por categoría

```python
for nombre, df in tiendas.items():
    print(f"\n📊 Categorías más vendidas en {nombre}:")
    print(df["Categoría del Producto"].value_counts().head(5))
```

# 3. Calificación promedio de la tienda


```python
calificaciones = {nombre: df["Calificación"].mean() for nombre, df in tiendas.items()}
calif_df = pd.DataFrame(list(calificaciones.items()), columns=["Tienda", "Calificación Promedio"])

sns.barplot(data=calif_df, x="Tienda", y="Calificación Promedio")
plt.title("Calificación promedio por tienda")
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.show()
```

# 4. Productos más y menos vendidos

```python
for nombre, df in tiendas.items():
    print(f"\n🔥 Productos más vendidos en {nombre}:")
    print(df["Producto"].value_counts().head(3))

    print(f"\n❄️ Productos menos vendidos en {nombre}:")
    print(df["Producto"].value_counts().tail(3))
```

# 5. Envío promedio por tienda

```python
envios = {nombre: df["Costo de envío"].mean() for nombre, df in tiendas.items()}
envios_df = pd.DataFrame(list(envios.items()), columns=["Tienda", "Costo Envío Promedio"])

sns.barplot(data=envios_df, x="Tienda", y="Costo Envío Promedio")
plt.title("Costo promedio de envío por tienda")
plt.xticks(rotation=45)
plt.show()
```
